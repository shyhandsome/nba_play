```python
###General Mapping Function###
#this function will read a lookup table and do the mapping for you, so you don't need to write classification codes every time.
#the output is your main data with the mapped column

#the function will require the following 5 inputs:

#Main_data <- your main dataset (type: dataframe)
#Lookup_table <- your lookup table (type: dataframe)
#left_name <- the key in your main dataset (type: string)
#right_name <- the key in your lookup table (type: string)
#result_column_name <- which column do you want to put the results (type: string)

def Mapping(Main_data, Lookup_table, left_name, right_name, result_coloumn_name):
    df1 = Main_data
    df2 = Lookup_table
    results = df1.merge(df2, left_on = left_name, right_on = right_name, how = 'left')
    result_column = results.iloc[:,-1]
    Main_data[result_coloumn_name] = result_column
```

```python
###Create Lookup Table Function###
#this function will create a lookup table as dataframe for you

#the function will require the following 3 inputs:

#Lookup_file <- your lookup file location (type: string)
#sheet_name <- your sheet name in the file
#n <- which column you want to return
def create_Lookup_table(Lookup_file, sheet_name, n):
    df = pd.read_excel(Lookup_file,sheet_name = sheet_name)
    df = df.iloc[:,[0,n-1]]
    return df
```

```python
def get_weekend_date(date):
    weekday_num = datetime.strptime(date, "%m/%d/%Y").date().weekday()+1
    MSFT_weekday_num = weekday_num%7
    week_end_date = (datetime.strptime(date, "%m/%d/%Y").date() + timedelta(6 - MSFT_weekday_num)).strftime('%m/%d/%Y')
    week_end_date_out = datetime.strptime(week_end_date, "%m/%d/%Y")
    return week_end_date_out
```
